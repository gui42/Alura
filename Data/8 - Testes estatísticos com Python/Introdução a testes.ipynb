```python
ax = sns.distplot(notas1['rating'], bins=10)
ax.figure.set_size_inches(16,4)
ax.set_title('Distribuição de Notas do Filme Toy Story',fontsize=20)
ax.set_ylabel('Densidade',fontsize=16)
ax.set_xlabel("Nota")
```

```python
sns.set_style('whitegrid')
ax = sns.boxplot(x=notas1['rating'])
ax.figure.set_size_inches(16,4)
ax.set_title("Distribuição das notas do filme Toy Story",fontsize=20)
ax.set_xlabel('Nota', fontsize=16)
```

```python
notas1.shape[0]
```

```python
zconfint(notas1['rating'])
```

```python
from statsmodels.stats.weightstats import ztest
np.random.seed(75241)
temp = notas1.sample(frac=1)['rating']

def calcula_test(i):
    media = temp[0:i].mean()
    stat,p = ztest(temp[0:i], value=3.4320)
    return (i, media, p)

medias=np.array([calcula_test(i) for i in range(2, temp.shape[0])])
```

```python
plt.plot(medias[:,1])
plt.plot(medias[:,2])
plt.hlines(y=0.05, xmin=2, xmax=temp.shape[0], colors='red')
```

```python
zconfint(notas1.rating, notas.rating)
```

```python
from scipy.stats import ttest_ind
ttest_ind(notas.rating, notas1.rating)
```

```python
descr_todas_as_notas = DescrStatsW(notas.rating)
descr_toystory_notas = DescrStatsW(notas1.rating)
comparacao = descr_todas_as_notas.get_compare(descr_toystory_notas)
comparacao.summary()
```

```python
plt.boxplot([notas.rating, notas1.rating], labels=['Todos os Filmes', 'Toy Story'])
plt.title('Distribuição das notas de acordo com filmes')
```

Teste com amostra menor

```python
plt.boxplot([notas.rating, notas1[3:12].rating], labels=['Todos os Filmes', 'Toy Story (do 3 ao 12)'])
plt.title('Distribuição das notas de acordo com filmes')
```

```python
descr_todas_as_notas = DescrStatsW(notas.rating)
descr_toystory_notas = DescrStatsW(notas1[3:12].rating)
comparacao = descr_todas_as_notas.get_compare(descr_toystory_notas)
comparacao.summary(use_t=True)
```

## Aula 6 - Não paramétricos
### Comparar a média de dois filmes

```python
filmes.query('movieId in [1, 593, 72226]')
```

```python
notas293 = notas.query('movieId == 593')
notas72226 = notas.query('movieId == 72226')
```

```python
plt.boxplot([notas1['rating'], notas293['rating'], notas72226['rating']], labels=['Toy Story',
                                                                                 'Silence Of the Lambs',
                                                                                 'Fantastic Mr. Fox'])
plt.title('Comparação de notas de alguns filmes')
```

```python
ax = sns.boxplot(data=notas.query('movieId in (1, 593, 72226)'), x='movieId', y='rating',)
ax.figure.set_size_inches(16,10)
ax.set_title('Boxplot das notas de algums filmes', fontsize=20)
ax.set_xlabel('Filme', fontsize=16)
```

```python
from scipy.stats import normaltest
stat, p = normaltest(notas1['rating'])
p
```

```python
from scipy.stats import ranksums
_, p = ranksums(notas1['rating'], notas293['rating'])
p
```

O ranksums testa distribuições não paramétricas, a interpretação do teste diz que uma nota aleatória do filme 293 tem maiores chances de ser mais alta do que uma nota aleatória do filme 1
